```python
from agents.v7.a2c_agent import A2CAgent
from utils.board import random_kqk_mate_in_one
```

```python
agent = A2CAgent(lr=3e-4, curriculum_ratio=1.0, entropy_coef=0.05)

for _ in range(5):
    agent.mate_pool = [random_kqk_mate_in_one().fen() for _ in range(500)]
    agent.train(n_episodes=100_000, gamma=0.99, log_every=25_000, n_workers=31, episodes_per_worker=16)
    agent.evaluate_mate_in_one(n_episodes=500, greedy=True)
```
