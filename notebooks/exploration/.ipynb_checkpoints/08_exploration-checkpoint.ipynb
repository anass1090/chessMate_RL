```python
from agents.v7.reinforce_agent import ReinforceAgent
```

```python
agent = ReinforceAgent(curriculum_ratio=1.0)

agent.train(n_episodes=500_000, log_every=50_000)
agent.evaluate_mate_in_one(n_episodes=500, greedy=True)
```

```python
agent.evaluate(n_episodes=10000)
```

```python
agent.save('kqk_reinforce_v7')
```

```python
for ratio in [0.8, 0.6, 0.4, 0.2, 0.0]:
    agent.curriculum_ratio = ratio
    agent.train(n_episodes=500_000)
```

```python
agent.evaluate_mate_in_one(n_episodes=500, greedy=True)
```

```python
agent.evaluate(n_episodes=10000)
```

```python
agent.save("kqk_reinforce_v7_stage_2")
```

```python
agent2 = ReinforceAgent()
agent2.load('kqk_reinforce_v7_stage_2')
```
