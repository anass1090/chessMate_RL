```python
from agents.v9.ppo_agent import PPOAgent
```

```python
agent = PPOAgent(curriculum_ratio=1.0)

agent.train(n_episodes=500_000, movement="centrum")
agent.evaluate_mate_in_one(n_episodes=500)
```

```python
agent.save('kqk_ppo_v1')
```

```python
for ratio in [0.8, 0.6, 0.4, 0.2, 0.0]:
    agent.curriculum_ratio = ratio
    agent.train(n_episodes=500_000)
    agent.evaluate_mate_in_one(n_episodes=500, greedy=True)
    agent.evaluate(n_episodes=500)
```

```python
agent.evaluate(n_episodes=10000)
```
